# Audio Feature Analysis

```python
import pandas as pd
import numpy as np
import altair as alt
from sklearn import preprocessing
from altair_saver import save
```

Disabling max row restriction from altair

```python
alt.data_transformers.disable_max_rows()
```

#### Importing Data

```python
cty = pd.read_csv('../Data/df_cty_final.csv')
rb = pd.read_csv('../Data/df_rb_final.csv')
rock = pd.read_csv('../Data/df_rock_final.csv')
```

```python
cty = cty[[ 'dance', 'energy','loud', 'speech', 'acoust', 'live', 'valence', 'tempo']]
cty = pd.DataFrame(preprocessing.minmax_scale(cty), columns = [ 'dance', 'energy','loud', 'speech', 'acoust', 'live', 'valence', 'tempo'])
```

```python
rb = rb[[ 'dance', 'energy','loud', 'speech', 'acoust', 'live', 'valence', 'tempo']]
rb = pd.DataFrame(preprocessing.minmax_scale(rb), columns = [ 'dance', 'energy','loud', 'speech', 'acoust', 'live', 'valence', 'tempo'])
```

```python
rock = rock[[ 'dance', 'energy','loud', 'speech', 'acoust', 'live', 'valence', 'tempo']]
rock = pd.DataFrame(preprocessing.minmax_scale(rock), columns = [ 'dance', 'energy','loud', 'speech', 'acoust', 'live', 'valence', 'tempo'])
```

#### Converting my data from wide form, to long form to be interpreted by alt air 

```python
cty = pd.melt(cty)
rb = pd.melt(rb)
rock = pd.melt(rock)
```

# Boxplots of Audio Feautres

```python
rock_chart = alt.Chart(rock, title = 'Rock Audio Features').mark_boxplot(opacity = .5,color = '#a24857').encode(
    x='value:Q',
    y='variable'
).properties(width = 900, height = 200)

rb_chart = alt.Chart(rb, title = 'Hip-Hop/Rap Audio Features').mark_boxplot(opacity = .5).encode(
    x='value:Q',
    y='variable'
).properties(width = 900, height = 200)

cty_chart = alt.Chart(cty, title = 'Country Audio Features').mark_boxplot(opacity = .5,color = '#D2B48C').encode(
    x='value:Q',
    y='variable'
).properties(width = 900, height = 200)
```

```python
af = alt.vconcat(rock_chart,rb_chart,cty_chart)
```
